```python
# Plot striplot (swarm plot wouldn't work) and violin plot pairs for all catagory varaibles with more than two possible values
columns = ['name_income_type', 'name_education_type', 'name_housing_type', 'occupation_type', '0-29', '30-59', '60-89', '90-119', '120-149', 'paid_off', 'no_debt']
for col in columns:
    plt.figure(figsize=(30,20))
    plt.subplot(211)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.xlabel(col, fontsize=18)
    plt.ylabel('Income', fontsize=18)
    plt.title(f'Income by {col}', fontsize=20)
    sns.stripplot(data=train, x=col, y='amt_income_total')
    plt.subplot(212)
    sns.violinplot(data=train, x=col, y='amt_income_total')
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.xlabel(col, fontsize=18)
    plt.ylabel('Income', fontsize=18)
    plt.title(f'Income by {col}', fontsize=20)
```
