```python
app.rename(columns={'CODE_GENDER': 'gender', 'FLAG_OWN_CAR': 'car_owner', 'FLAG_OWN_REALTY': 'property_owner', 'CNT_CHILDREN': 'children_count', 'AMT_INCOME_TOTAL': 'annual_income', 'NAME_INCOME_TYPE': 'income_type', 'NAME_EDUCATION_TYPE': 'education'}, inplace=True)
```

```python
app.head(1)
```

```python
app.rename(columns={'NAME_FAMILY_STATUS': 'family_status', 'NAME_HOUSING_TYPE': 'housing_type', 'DAYS_BIRTH': 'days_from_birth', 'DAYS_EMPLOYED': 'days_employed', 'FLAG_MOBIL': 'mobile', 'FLAG_WORK_PHONE': 'work_phone', 'FLAG_PHONE': 'phone', 'FLAG_EMAIL': 'email', 'OCCUPATION_TYPE': 'occupation', 'CNT_FAM_MEMBERS': 'family_member_count'  }, inplace=True)
```

```python
app.head(1)
```

```python
app['days_from_birth'] = abs(app.days_from_birth)
```

```python
# app['days_employed'] = abs(app.days_employed)
```

```python
app.head(1)
```

```python
app['car_owner'] = app['car_owner'].map({'Y': 1, 'N': 0})
```

```python
app['property_owner'] = app['property_owner'].map({'Y': 1, 'N': 0})
```

```python
app.head(1)
```

```python
#Changing float columns to integers
cols = ['annual_income', 'family_member_count']
for col in cols:
   app[col] = app[col].apply(lambda x: int(x))
```

```python
app.head(1)
```

```python
main = app.merge(df, on='ID', how='left')
```

```python
main.head()
```

```python
main.rename(columns={'MONTHS_BALANCE': 'months_balance', 'STATUS': 'status'}, inplace=True)
```

```python
main['months_balance'] = abs(main.months_balance)
```
